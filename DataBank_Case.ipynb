Achei interessante trazer um glossário para o que significa algumas dessas colunas, com objetivo de consultas no decorrer da análise. Muitas vezes a coluna vem nomeada de uma forma vaga, então é necessário consultar a documentação do dataset para validar o conteúdo.

* default - possui valor inadimplente?
* balance - média de saldo anual
* housing - possui financiamento de habitação?
* campaign - número de contatos realizados com esse cliente na campanha atual
* pdays - número de dias desde o último contato com o cliente na última campanha, se for -1 quer dizer que o cliente ainda não foi contactado
* previous - número de contatos performados antes da campanha atual
* poutcome - resultado da campanha anterior, variável categórica
* y - o cliente adquiriu o CDB?
